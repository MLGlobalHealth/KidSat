```python
import re
import matplotlib.pyplot as plt

def plot_losses(filenames):
    plt.figure(figsize=(10, 5))  # Set up the figure for plotting

    # Iterate over each file
    for filename in filenames:
        # Initialize lists to store epochs and individual losses
        epochs = []
        individual_losses = []

        # Regular expression to capture the required data
        pattern = r"Epoch \[(\d+)/\d+\], Validation Loss: [-\d.]+, Individual Loss: tensor\(\[([-.\d]+)\]"

        # Open and parse the log file
        with open(filename, "r") as file:
            for line in file:
                match = re.search(pattern, line)
                if match:
                    # Extract epoch and individual loss
                    epoch = int(match.group(1))
                    individual_loss = float(match.group(2))
                    epochs.append(epoch)
                    individual_losses.append(individual_loss)

        # Plot each file's individual loss with a unique label
        plt.plot(epochs, individual_losses, marker='o', linestyle='-', label=filename)

    # Customize the plot
    plt.title("Individual Loss over Epochs for Multiple Files")
    plt.xlabel("Epoch")
    plt.ylabel("Individual Loss")
    # plt.xlim(0, 20)  # Set the x-axis limits
    plt.legend(['SWIN lr=1e-5','ViT lr=1e-5','DINO','SWIN lr=1e-4','ViT lr=1e-4',],loc="upper right")  # Add a legend to distinguish different files
    plt.grid(True)
    plt.show()

plot_losses(['submit_slurm/spatial_swin1S_deprived_sev.log','submit_slurm/spatial_vits1S_deprived_sev.log','submit_slurm/spatial1S_deprived_sev.log',
             'submit_slurm/spatial_swin1S_deprived_sev_1e4.log','submit_slurm/spatial_vits1S_deprived_sev1e4.log'])
```
